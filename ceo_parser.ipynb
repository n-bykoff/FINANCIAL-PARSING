```python
from selenium import webdriver
import pandas as pd
import numpy as np
import time
```

```python
# function for logging in to finviz.com 
def login_on_finviz():
    wb = webdriver.Chrome()
    wb.get('https://finviz.com/login.ashx')

    inputs = wb.find_elements_by_class_name('input')
    inputs[0].send_keys('Your Login')
    inputs[1].send_keys('Your Password')

    btn = wb.find_element_by_class_name('button')

    time.sleep(1)
    btn.click()
    time.sleep(1.5)
    
    return wb
```

```python
# function to go to the desired page by the ticker
def go_to_company_page(wb, ticker):
    wb.get(f'https://elite.finviz.com/quote.ashx?t={ticker}')
    time.sleep(5)
    
    return wb
```

```python
def get_ceo_info(wb, ticker):
    
    wb = go_to_company_page(wb, ticker)
    a = wb.find_elements_by_class_name('tab-link')
    
    for el in a:
        if el.text == '...more':
            url = el.get_attribute("href")
    
    wb.get(url)
    
    with open(f'./CEO/{ticker}.txt', 'w') as f:    
        header = wb.find_element_by_class_name('MarketsTable-header-2NLA1')
        f.write(header.text + '\n')
        
        data = wb.find_elements_by_class_name('data')
        for line in data:
            f.write(line.text + '\n')
        
    return wb
```

```python
data = pd.read_excel('Screener tests.xlsx')
top_data = data[data['Cap M$'] != 'No data on finviz'][:100]
tickers = list(top_data['Все'])
```

```python
wb = login_on_finviz()

for ticker in tickers:
    try:
        wb = get_ceo_info(wb, ticker)
        print(f'{ticker} - DONE!')
    except:
        print(f'{ticker} - ошибка!')
```

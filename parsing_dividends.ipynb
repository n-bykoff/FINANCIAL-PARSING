```python
data = pd.read_excel('./data/Screener tests.xlsx')
data = data[data['Cap M$'] != 'No data on finviz']
tickers = list(data['Все'])
```

```python
df_dic = {}
wd = webdriver.Chrome()

for ticker in tickers[94:]:
    try:
        wd = go_to_search_page(wd)
        text = get_text(wd, ticker)
        div_dic = reshape_div_dic(make_div_dic(text))
        df_dic[ticker] = make_df(div_dic)
        df_dic[ticker].to_excel(f'./data/dividends history/{ticker}.xlsx')
        print(ticker, '- DONE!')
    except:
        print(ticker, '- ERROR!')
        
wd.close()
```
